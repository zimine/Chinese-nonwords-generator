```python
df_pho = pd.read_csv("/Users/zimine/Documents/MPhil/Stimuli/Data/CLP_pho.csv")
df_pho
```

```python
#Creat four columns to df_clp and set all values to 0
df_clp['HD-1']=0
df_clp['HD-2']=0
df_clp['Num-neighbor-1']=0
df_clp['Num-neighbor-2']=0
df1 = df_clp.drop(columns=['C&B-Subtitle-CD-C1','C&B-Subtitle-CD-C2','C&B-Subtitle-CD-W'])
df1
```

```python
#Get the index values of df_pho
index_pho = df_pho.index.values

#Convert the index array to list
index_pho_new = index_pho.tolist()

len(index_pho_new)
```

```python
#a = [None] * 4874
#len(a)
```

```python
#Check a raondom value
any(df1['Character-1'].str.contains(df_pho['Character'][2])) == True
```

```python
#Check how many rows met the criteria
a = df1.index.values[df1['Character-1'].str.contains(df_pho['Character'][2]) == True]
a
```

```python
#Search in df1 to look for every character in df_pho
#Mutiple rows in df1 may contain same character in df_pho
#Record the homodensity of every character found in df1
#Fill the homodensity of every character in df_pho


for i in index_pho_new: 
    if any(df1['Character-1'].str.contains(df_pho['Character'][i])) == True:
        a = df1.index.values[df1['Character-1'].str.contains(df_pho['Character'][i]) == True]
        for k in a:
            m = df_pho.at[i,'HomoDensity']
            df1.at[k,'HD-1'] = m
    else:
        df1.at[i, 'HD-1'] = -1
```

```python
#Check df1
df1
```

```python
#Look for rows when HD-1 equals 0
#These characters are missing in df_pho
#Set the HD-1 value of these rows to None.

df2 = df1.loc[df1['HD-1'] == 0]
df2.loc[df2['HD-1'] == 0, 'HD-1'] = ''
df2
```

```python
#Update df1 by df2, which means update by index, and rows with same index value will be updated by df2 

df1.update(df2)
df1
```

```python
#Look for rows when HD-1 equals -1

df3 = df1.loc[df1['HD-1'] == -1]
df3
```

```python
#Put the df 3 in to for loop again to check 

for i in index_pho_new: 
    if any(df3['Character-1'].str.contains(df_pho['Character'][i])) == True:
        a = df3.index.values[df3['Character-1'].str.contains(df_pho['Character'][i]) == True]
        for k in a:
            m = df_pho.at[i,'HomoDensity']
            df3.at[k,'HD-1'] = m
    else:
        df3.at[i, 'HD-1'] = -1
```

```python
df3
```

```python
#Set the HD-1 value to rows still missing after check
#Convert df3 into csv for cannot delet the last row 

df3.loc[df3['HD-1'] == -1, 'HD-1'] = ''
df3.loc[df3['HD-1'] == '']
df3.to_csv('df3.csv')
```
