```python
# get the length of the string of column in a dataframe

df['word_length'] = df['word'].apply(len)

# Select rows from a DataFrame based on values in a column 

df.append = df.loc[df['word_length'] == 1]
df.append = df.append.loc[df.append['frequency'] >= 3]
```

```python
word_list = df.append["word"].tolist()
word_list
```

```python
word_string = "".join(word_list)
word_string
```

```python
#Print out all two-character pseudoword permutations

list(permutations(word_list, 2))

#Print out all two-character pseudoword combinations

pseudoword_list = list(combinations(word_list, 2))
```
