```python
#-------------------------8th step--------------------------
df1 = pd.read_csv("/Users/zimine/Documents/MPhil/Stimuli/Data/df1_ver_2.csv", index_col = 0)
df1
```

```python
df_pho = pd.read_csv("/Users/zimine/Documents/MPhil/Stimuli/Data/CLP_pho.csv")
df_pho
```

```python
#Get the index values of df_pho
index_pho = df_pho.index.values

#Convert the index array to list
index_pho_new = index_pho.tolist()

len(index_pho_new)
```

```python
#Search in df1 to look for every character in df_pho
#Mutiple rows in df1 may contain same character in df_pho
#Record the number of neighbours of every character found in df1
#Fill the number of neighbours of every character in df_pho

for i in index_pho_new: 
    if any(df1['Character-2'].str.contains(df_pho['Character'][i])) == True:
        a = df1.index.values[df1['Character-2'].str.contains(df_pho['Character'][i]) == True]
        for k in a:
            m = df_pho.at[i,'Num-Neighbors']
            df1.at[k,'Num-neighbor-2'] = m
    else:
        df1.at[i, 'Num-neighbor-2'] = -1
df1
```

```python
#Check rows when Num-neighbor-2 equals to -1 and 0

df2 = df1.loc[df1['Num-neighbor-2'] == -1]
```

```python
#Search in df1 to look for every character in df_pho
#Mutiple rows in df1 may contain same character in df_pho
#Record the number of neighbours of every character found in df1
#Fill the number of neighbours of every character in df_pho

for i in index_pho_new: 
    if any(df2['Character-2'].str.contains(df_pho['Character'][i])) == True:
        a = df2.index.values[df2['Character-2'].str.contains(df_pho['Character'][i]) == True]
        for k in a:
            m = df_pho.at[i,'Num-Neighbors']
            df2.at[k,'Num-neighbor-2'] = m
    else:
        df2.at[i, 'Num-neighbor-2'] = -1
df2
```

```python
#Convert Num-neighbor-2 of character missing in df_pho to None
#Delete last row and read df2

df2.loc[df2['Num-neighbor-2'] == -1, 'Num-neighbor-2'] = ''
df2.to_csv('df2_ver_2.csv')
```
